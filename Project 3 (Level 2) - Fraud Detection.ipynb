```python
X = pd.DataFrame(data.iloc[:,:-1])
y = pd.DataFrame(data.iloc[:,-1]) 
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()

regressor.fit(X_train, y_train)
```

```python
y_pred_train = regressor.predict(X_train)
```

```python
y_pred_test = regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score_train = accuracy_score(y_pred_train, y_train)
accuracy_score_test = accuracy_score(y_pred_test, y_test)
```

```python
print("Accuracy on training data: ", accuracy_score_train)
print("Accuracy on testing data: ", accuracy_score_test)
```
